```python
INDIC_NLP_RESOURCES=r"/content/drive/MyDrive/Colab Notebooks/L2C4/indic_nlp_resources-master"

from indicnlp.transliterate.unicode_transliterate import ItransTransliterator

from indicnlp import common, loader
common.set_resources_path(INDIC_NLP_RESOURCES)
loader.load()
```

```python
a = df['manually_translated_hindi'].tolist()
```

```python
lang='hi'
import numpy as np
romanised_hi_keys = []

for key in tqdm(a):
    print(key)
    try: r = ItransTransliterator.to_itrans(key,lang)
    except: r = ''
    r = r.replace("j~n", "gy").replace(".", "").replace("^", "").replace("ॉ", "o").replace("a़", "on").replace("ऑ", "o").replace("ँ", "n").replace('aa', 'a').replace('ii', 'i').replace('ee', 'e').replace('uu', 'u').lower()
    #r = "".join([i + ' ' for i in [i[:-1] if i[-1] == 'a' else i for i in r.split()]])
    print(r)
    print('--------------------------------------------------')
    romanised_hi_keys.append(r)
```

```python
len(romanised_hi_keys)
```

```python
df['translit'] = romanised_hi_keys
df
```
